### Gráficos no python - Seaborn

Voltada para parte estatística

```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

```

```python
df = pd.read_excel('videosYT.xlsx','videos')
display(df.head())
```

1. relplot(relational) -> scatterplot, lineplot
2. displot(distributions) -> histplot, kdeplot, ecdfplot, rugplot
3. catplot(categorical) -> stripplot, swarmplot, boxplot, violinplot, pointplot, barplot

### Gráfico de dispersão

```python
fig = sns.scatterplot(data=df, x='Nº de Views', y='Nº de Likes')
```

```python
fig = sns.scatterplot(data=df, x='Nº de Views', y='Nº de Likes', hue='Responsável', style='Responsável')
```

### Gráfico de dispersão relacional

```python
fig_rel = sns.relplot(data=df, x='Nº de Views', y='Nº de Likes', hue='Categoria', col='Responsável')
```

### Gráficos de linha

```python
dataset = pd.read_excel('videosYT.xlsx', 'Inscritos')
display(dataset.head())
display(dataset.shape)
```

```python
fig_line = sns.lineplot(data=dataset, x='Mês/Ano', y='Inscritos')
plt.xticks(rotation=45)
```

### Criando histogramas

```python
fig_hist= sns.displot(data=df, x='Nº de Views', kind='hist') #por padrão ele traz o histplot
```

```python
fig_hist= sns.displot(data=df, x='Nº de Views', kind='kde')
```

Esse gráfico trás o lugar onde está mais concentrado a informação, ou seja, a região com maior densidade

```python
fig_hist= sns.displot(data=df, x='Nº de Views', kind='ecdf')
```

```python
fig_hist= sns.displot(data=df, x='Nº de Views', kind='hist', hue='Responsável')
```

```python
fig_hist= sns.displot(data=df, x='Nº de Views', kind='hist', hue='Responsável', col='Responsável')
plt.show(fig_hist)
```

### Categorical plots


```python
fig_hist= sns.lmplot(data=df, x='Nº de Views',y='Nº de Likes', hue='Responsável', markers=['o','x'])
plt.show(fig_hist)
```

```python
fig_hist= sns.regplot(data=df, x='Nº de Views',y='Nº de Likes')
plt.show(fig_hist)
```
